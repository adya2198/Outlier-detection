```python
import xml.etree.ElementTree as ET
from termcolor import colored

tree = ET.parse('DataSet1 (1).xml')
root = tree.getroot()

# Print the root element and its children

for child in root:
    if child.tag=='Products':
#IceCat_Level2------------------------------------------------------------------------------------------------------------------
        for level2 in child:
            if level2.tag == 'Product':
                print('->',level2.find('Name').text)
#IceCat_Level3------------------------------------------------------------------------------------------------------------------
                for level3 in level2:
                    if level3.tag == 'Product':
                        print('-->',level3.find('Name').text)
#IceCat_Level4------------------------------------------------------------------------------------------------------------------
                        for level4 in level3:
                            if level4.tag == 'Product':
                                print('---->',level4.find('Name').text)
                                list0=[]
                                attrib_links = level4.findall("AttributeLink")
                                for attrib_link in attrib_links:
                                    attribute_id = attrib_link.get("AttributeID")
                                    list0.append(attribute_id)  
                                if len(list0) != 0:
                                    for kk in level4:
                                        values = kk.find('Values')
                                        if values is not None:
                                            for value_elem in values:
                                                attri=value_elem.get("AttributeID")
                                                if attri != "AT_AIAC_product_description":
                                                    flag2=False
                                                    for list_elem in list0:
                                                        if attri == list_elem:
                                                            flag2=True
                                                    if flag2 == False:
                                                        print('---->',level4.find('Name').text)
                                                        s="*******************"+kk.find('Name').text
                                                        print(colored(s,'red'))
                                                        break
#IceCat_Level5------------------------------------------------------------------------------------------------------------------
                                for level5 in level4:
                                    if level5.tag=='Product':
                                        list1=[]
                                        attrib_links = level5.findall("AttributeLink")
                                        for attrib_link in attrib_links:
                                            attribute_id = attrib_link.get("AttributeID")
                                            list1.append(attribute_id)   
                                        if len(list1) != 0:
                                            for ll in level5:
                                                values = ll.find('Values')
                                                if values is not None:
                                                    for value_elem in values:
                                                        attri=value_elem.get("AttributeID")
                                                        if attri != "AT_AIAC_product_description":
                                                            flag2=False
                                                            for list_elem in list1:
                                                                if attri == list_elem:
                                                                    flag2=True
                                                            if flag2 == False:
                                                                print("-------->",level5.find('Name').text)
                                                                s="*******************"+ll.find('Name').text
                                                                print(colored(s,'red'))
                                                                break
#IceCat_Level6------------------------------------------------------------------------------------------------------------------
                                        for level6 in level5:
                                            if level6.tag=='Product':
                                                list2=[]
                                                attrib_links = level6.findall("AttributeLink")
                                                for attrib_link in attrib_links:
                                                    attribute_id = attrib_link.get("AttributeID")
                                                    list2.append(attribute_id)   
                                                if len(list2) != 0:
                                                    for mm in level6:
                                                        values = mm.find('Values')
                                                        if values is not None:
                                                            for value_elem in values:
                                                                attri=value_elem.get("AttributeID")
                                                                if attri != "AT_AIAC_product_description":
                                                                    flag2=False
                                                                    for list_elem in list2:
                                                                        if attri == list_elem:
                                                                            flag2=True
                                                                    if flag2 == False:
                                                                        print("-------->",level6.find('Name').text)
                                                                        s="*******************"+mm.find('Name').text
                                                                        print(colored(s,'red'))
                                                                        break
#IceCat_Level7------------------------------------------------------------------------------------------------------------------
                                                for level7 in level6:
                                                    if level7.tag=='Product':
                                                        list3=[]
                                                        attrib_links = level7.findall("AttributeLink")
                                                        for attrib_link in attrib_links:
                                                            attribute_id = attrib_link.get("AttributeID")
                                                            list3.append(attribute_id)   
                                                        if len(list3) != 0:
                                                            for nn in level7:
                                                                values = nn.find('Values')
                                                                if values is not None:
                                                                    for value_elem in values:
                                                                        attri=value_elem.get("AttributeID")
                                                                        if attri != "AT_AIAC_product_description":
                                                                            flag2=False
                                                                            for list_elem in list3:
                                                                                if attri == list_elem:
                                                                                    flag2=True
                                                                            if flag2 == False:
                                                                                print("-------->",level7.find('Name').text)
                                                                                s="*******************"+nn.find('Name').text
                                                                                print(s,'red')
                                                                                break
                                     
```
